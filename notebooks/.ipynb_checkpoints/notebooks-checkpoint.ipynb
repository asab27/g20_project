```python
import sqlite3
import pandas as pd
```

```python
csv_path = "../G20_Tech Support â€“ Tickets  Agents with Solutions_merged.csv"
df = pd.read_csv(csv_path, delimiter=';', encoding='utf-8')
df.head()
```

```python
conn = sqlite3.connect("../data/tech_support.db")
cursor = conn.cursor()
```

```python
cursor.execute('''CREATE TABLE IF NOT EXISTS Ticket (
                    ticket_id INTEGER PRIMARY KEY,
                    issue TEXT,
                    submission_date TEXT,
                    status TEXT)''')

cursor.execute('''CREATE TABLE IF NOT EXISTS Agent (
                    agent_id INTEGER PRIMARY KEY,
                    name TEXT,
                    department TEXT,
                    email TEXT)''')

cursor.execute('''CREATE TABLE IF NOT EXISTS TicketAssignment (
                    ticket_id INTEGER,
                    agent_id INTEGER,
                    assignment_date TEXT,
                    resolution_time INTEGER,
                    FOREIGN KEY(ticket_id) REFERENCES Ticket(ticket_id),
                    FOREIGN KEY(agent_id) REFERENCES Agent(agent_id))''')

cursor.execute('''CREATE TABLE IF NOT EXISTS Solution (
                    solution_id INTEGER PRIMARY KEY,
                    ticket_id INTEGER UNIQUE,
                    resolution_summary TEXT,
                    solved_date TEXT,
                    FOREIGN KEY(ticket_id) REFERENCES Ticket(ticket_id))''')

conn.commit()
print("Base de dados e tabelas criadas com sucesso!")
```

```python
for _, row in df.iterrows():
    cursor.execute("INSERT OR IGNORE INTO Ticket VALUES (?, ?, ?, ?)", 
                   (row['tickets_id'], row['issue'], row['submission_date'], row['status']))
    
    cursor.execute("INSERT OR IGNORE INTO Agent VALUES (?, ?, ?, ?)", 
                   (row['agents_id'], row['agents_name'], row['agents_department'], row['email']))
    
    cursor.execute("INSERT INTO TicketAssignment VALUES (?, ?, ?, ?)", 
                   (row['tickets_id'], row['agents_id'], row['assignment_date'], row['resolution_time']))
    
    cursor.execute("INSERT OR IGNORE INTO Solution VALUES (?, ?, ?, ?)", 
                   (row['solutions_id'], row['tickets_id'], row['resolution_summary'], row['solved_date']))

conn.commit()
conn.close()

print("Dados do CSV importados com sucesso para a base de dados!")
```
